## Constructing clonal consensus sequences

````python
CM0 = '/data/personal/jiewen.du/CM_new/data/clones/CM0_train_clone-pass_germ-pass_mutation.tsv'
CM1 = '/data/personal/jiewen.du/CM_new/data/clones/CM1_valid_clone-pass_germ-pass_mutation.tsv'
CM12 = '/data/personal/jiewen.du/CM_new/data/clones/CM12_valid_clone-pass_germ-pass_mutation.tsv'
CM13 = '/data/personal/jiewen.du/CM_new/data/clones/CM13_valid_clone-pass_germ-pass_mutation.tsv'
CM5_2= '/data/personal/jiewen.du/CM_new/data/clones/CM5_2_valid_clone-pass_germ-pass_mutation.tsv'
CM20= '/data/personal/jiewen.du/CM_new/data/clones/CM20_valid_clone-pass_germ-pass_mutation.tsv'
````

````python
for (i in c(1,2,3,4,'5_2',6,7,8,10,12,13,15,20,21)){ # 
    CM = paste('/data/personal/jiewen.du/CM_new/data/clones/CM', i, '_train_clone-pass_germ-pass_merged.tsv', sep='')
    df <- read.table(file=CM,
                  header=TRUE,
                  sep='\t')
    df_clones <- collapseClones(df, 
                     cloneColumn="clone_id", 
                     sequenceColumn="sequence_alignment", 
                     germlineColumn="germline_alignment_d_mask", 
                     regionDefinition=IMGT_V, 
                     method="thresholdedFreq", 
                     minimumFrequency=0.5,
                     includeAmbiguous=FALSE,
                     breakTiesStochastic=FALSE,
                     nproc=1)
    df_observed <- observedMutations(df_clones, 
                              sequenceColumn="clonal_sequence",
                              germlineColumn="clonal_germline",
                              regionDefinition=IMGT_V, 
                              frequency=TRUE,
                              nproc=1)

    df_expected_MK_RS5NF <- expectedMutations(df_observed, 
                                  sequenceColumn="clonal_sequence",
                                  germlineColumn="clonal_germline",
                                  targetingModel=MK_RS5NF,
                                  regionDefinition=IMGT_V, nproc=1)
    ofile = paste('/data/personal/jiewen.du/CM_new/data/mutations/CM', i, '_train_MK_RS5NF.tsv', sep='')
    write.table(df_expected_MK_RS5NF, ofile, sep='\t', quote=FALSE, col.names=TRUE, row.names=FALSE)
}
````
